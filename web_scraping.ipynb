```python
import requests
from bs4 import BeautifulSoup
```

```python
response = requests.get('http://www.columbia.edu/~fdc/sample.html')
response
```

```python
page = BeautifulSoup(response.text, 'html.parser')
page
```

```python
page.title
```

```python
page.title.string
```

```python
page.find_all('h3')
```

```python
links_section = page.find('h3', attrs={'id': 'chars'})
section = []
```

```python
for element in links_section.next_elements:
    if element.name == 'h3':
        break
    
    section.append(element.string or '')
```

```python
result = ''.join(section)
```

```python
result
```

```python
import feedparser
import datetime
import delorean
import requests
```

```python
rss = feedparser.parse('http://rss.nytimes.com/services/xml/rss/nyt/HomePage.xml')
```

```python
rss.channel.updated
```

```python
time_limit = delorean.parse(rss.channel.updated) - datetime.timedelta(hours=6)
```

```python
entries = [entry for entry in rss.entries if delorean.parse(entry.published) > time_limit]
```

```python
len(entries)
```

```python
len(rss.entries)
```

```python
entries[6]['title']
```

```python
entries[6]['link']
```

```python
requests.get(entries[6].link)
```

<p>Using Selenium for advanced interaction</p>

```python
from selenium import webdriver
```

```python
browser = webdriver.Chrome()
browser.get('https://httpbin.org/forms/post')
```

```python
custname = browser.find_element(by='name', value='custname')
custname.clear()
custname.send_keys('Sean Cornell')
```

```python
for size_element in browser.find_element(by='name', value='size'):
    if size_element.get('value') == 'medium':
        size_element.click()
```
